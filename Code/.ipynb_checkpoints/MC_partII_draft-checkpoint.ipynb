```python
n_miRNA = 12
n_iter = 1000

def get_df_edges(path_miR_dict, miRname):
    pathes = path_miR_dict[miRname]
    miR_col = [miRname for i in range(len(pathes))]
    return pd.DataFrame(list(zip(miR_col, pathes)), columns=['miR', 'path'])

def grep_path_gready(degr_cnt):
    """
    degr_cnt - dict = {node: degree}
    """
    last_val = 0
    cnt_pathes = list()
    for key, val in degr_cnt.items():
        if 'hsa-' in key:
            continue
        if not last_val:
            last_val = val
            cnt_pathes.append(key)
            continue
        if val != last_val:
            break
        else:
            cnt_pathes.append(key)
    return cnt_pathes

def grep_neighbors_set(cnt_pathes, bi_g):
    """
    cnt_pathes - list of nodes
    bi_g - nx.graph
    """
    mirna_set = set()
    for path in cnt_pathes:
        mirna_set.update(list(bi_g.neighbors(path)))
    return mirna_set

spec_path = ['R-HSA-170834', 'R-HSA-2173793', 'R-HSA-2173796']
res_total = list()
res_special = list()

tis_gene_set = miRNET.tissue_selector(ans=0, tissue_id=23)

for j in range(n_iter):
    print('iter = ',j)

    np.random.seed(j)
    miR_set = np.random.choice(list(dict_for_mc.keys()), size=n_miRNA, replace=False)
    
    list_paths = list()
    list_target_paths = list()
    
    data_edges = pd.DataFrame(columns=['miR', 'path'])
    for miR in miR_set:
        key_genes = dict_for_mc[miR]['key_genes']  # key nodes extraction
        
        #enrichment
        data_edges = pd.concat([data_edges, get_df_edges(dict_miR_to_paths, miR)], axis=0)
        
    #bigraph constraction
    bi_g = nx.from_pandas_edgelist(data_edges, source='miR', target='path')
    if not nx.is_connected(bi_g):
        bi_g = sorted(nx.connected_component_subgraphs(bi_g), key=len, reverse=True)[0]
    
    print('nodes:', bi_g.number_of_nodes(), 'edges:', bi_g.number_of_edges())
    #degree calculation
    degr_cnt = dict(bi_g.degree)
    degr_cnt = dict(sorted(degr_cnt.items(), key=lambda item: item[1], reverse=True))
    
    #central pathways and theyr miRNAs gready extraction
    cent_pathways = grep_path_gready(degr_cnt)
    gready_mirnas = grep_neighbors_set(cent_pathways, bi_g)
    
    #number of miRNA targeted of central pathways
    res_total.append(len(gready_mirnas))
    
    print('cent_pathways:', len(cent_pathways), 'cent_mirnas:', len(gready_mirnas))
    #check if central patways contain special patways
    gready_mirnas_spec = set()
    for path in spec_path:
        if path in cent_pathways:
            #theyr miRNAs gready extraction
            gr_mir = grep_neighbors_set([path], bi_g)
            gready_mirnas_spec.update(gr_mir)
            
    #number of miRNAs targeted of central pathways if their is special       
    res_special.append(len(gready_mirnas_spec))
    print()
```
